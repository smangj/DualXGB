```python
import pandas as pd
```

```python
soybean_data = pd.read_csv("data/豆粕.csv")
soybean_data.head()
```

```python
soybean_data.tail(20)
```

```python
mission1 = soybean_data[["date", "close_futures", "log_ret_futures", "his_rv_1month_futures"]].dropna()
mission1.tail()
```

```python
mission1.to_csv("m1.csv")
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
```

```python
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 18), sharex=True)
axes[0].plot(mission1["date"], mission1["close_futures"])
axes[0].text(0.5, 0.9, '豆粕指数期货价格', transform=axes[0].transAxes, ha='center')
axes[0].set_ylabel('价格')
axes[1].plot(mission1["date"], mission1["log_ret_futures"])
axes[1].text(0.5, 0.9, '豆粕指数期货价格收益率', transform=axes[0].transAxes, ha='center')
axes[1].set_ylabel('收益率')
axes[2].plot(mission1["date"], mission1["his_rv_1month_futures"])
axes[2].text(0.5, 0.9, '豆粕指数期货波动率', transform=axes[0].transAxes, ha='center')
axes[2].set_ylabel('波动率')
axes[2].set_xlabel('日期')
# plt.xticks(rotation=45)
# plt.tight_layout() 
plt.show()
```
